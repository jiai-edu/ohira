# Titanicのデータでfeaturetoolsを試す

```python
import featuretools
import pandas as pd
```

```python
train = pd.read_csv('../input/train.csv')
train = train.drop('Survived', 1)
```

# EntitySetの準備をする

```python
es: featuretools.entityset.entityset.EntitySet

es = featuretools.EntitySet(id='data')

es.entity_from_dataframe(entity_id='train',
                         dataframe=train,
                         index='PassengerId')
```

# 利用可能なPrimitiveを見る

```python
p = featuretools.list_primitives()
```

```python
p[p['type'] == 'transform']
```

```python
# いじりやすいようにソートした
p[p['type'] == 'transform'].sort_values('name').name.values
```

# どんな特徴量が生成されるかを試す

```python
# コメント解除したりして試せるよ

trans = [
#'absolute',
 'add_numeric',
# 'add_numeric_scalar',
# 'and', # いれても意味無い
# 'cum_count',
# 'cum_max',
# 'cum_mean',
# 'cum_min',
# 'cum_sum',
# 'day',
# 'days_since',
# 'diff',
# 'divide_by_feature',
# 'divide_numeric',
# 'divide_numeric_scalar',
# 'equal',
# 'equal_scalar',
# 'greater_than',
# 'greater_than_equal_to',
# 'greater_than_equal_to_scalar',
# 'greater_than_scalar',
# 'haversine',
# 'hour',
# 'is_null',
# 'is_weekend',
# 'isin',
# 'latitude',
# 'less_than',
# 'less_than_equal_to',
# 'less_than_equal_to_scalar',
# 'less_than_scalar',
# 'longitude',
# 'minute',
# 'modulo_by_feature',
# 'modulo_numeric',
# 'modulo_numeric_scalar',
# 'month',
# 'multiply_numeric',
# 'multiply_numeric_scalar',
# 'negate',
# 'not',
# 'not_equal',
# 'not_equal_scalar',
# 'num_characters',
# 'num_words',
# 'or',
# 'percentile',
# 'scalar_subtract_numeric_feature',
# 'second',
# 'subtract_numeric',
# 'subtract_numeric_scalar',
# 'time_since',
# 'time_since_previous',
# 'week',
# 'weekday',
# 'year'
]


feature_names = featuretools.dfs(entityset=es,
                                 target_entity='train',
                                 verbose=1,
                                 agg_primitives=[],
                                 trans_primitives=trans,
                                 features_only=True,
                                 max_depth=1
                                 )
feature_names
```

# 実際に特徴量を作って、処理時間と結果を見る

```python
df, fs = featuretools.dfs(entityset=es,
                         target_entity='train',
                         verbose=1,
                         agg_primitives=[],
                         trans_primitives=trans,
                         features_only=False,
                         max_depth=1
                         )
```
