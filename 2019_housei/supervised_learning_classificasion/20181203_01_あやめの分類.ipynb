# あやめの分類

## データを知る

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
```

```python
# どういうデータなのかという情報がそもそも格納されている
print(iris_dataset.DESCR)
```

```python
# 特徴量は4種類
iris_dataset["feature_names"]
```

```python
# 4次元のnumpy配列が150ある
iris_dataset["data"]
```

```python
# データの形状は .shape でアクセスできる
iris_dataset["data"].shape
```

```python
# クラスは3種類
iris_dataset["target_names"]
```

```python
# 0, 1, 2 のいずれかの値でクラスを表現している
iris_dataset["target"]
```

```python
# 形状を確認
iris_dataset["target"].shape
```

## データを分割する

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset["data"], 
    iris_dataset["target"],
    random_state=0
)
```

```python
# 分割後の形状確認(学習データ)
X_train.shape, y_train.shape
```

```python
# 分割後の形状確認(テストデータ)
X_test.shape, y_test.shape
```

## データの観察
- いきなり機械学習モデル構築をするわけじゃあない


```python
import pandas as pd

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset["feature_names"])
```

```python
iris_dataframe.head()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

grr = pd.plotting.scatter_matrix(
    iris_dataframe,
    c=y_train,
    figsize=(15, 15),
    marker="o",
    s=60,
    alpha=0.8,
    cmap="Set1"
)
```

### 可視化してみてどうか？
- sepal(がく片) と petal(花弁) で結構分離できている気がする
- 品種を区別するための情報がこのデータに含まれていそう

## 機械学習モデルを構築
- 今回は最近傍法(k-Nearest-Neighbor)を適用
    - 現時点では手法は特にフォーカスしない
    - パラメータはとりあえず `n_neighbor=1`
    

```python
from sklearn.neighbors import  KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

## 予測する
- 新しいあやめの花で試してみよう

```python
import numpy as np

X_new = np.array([[5.0, 2.9, 1.0, 0.2]])
X_new.shape
```

```python
prediction = knn.predict(X_new)

prediction
```

### クラス0、つまり setosa であると判断したけど...?
- でもこれを信じてよいものか？
- このサンプルの本当の品種がわからないのならモデルを作っても意味がない


## モデルの評価
- テストデータを使って、構築した機械学習モデルの性能を評価する
    - テストデータ == 「モデル構築に利用していない」 かつ 「ラベルが既知」

```python
y_pred = knn.predict(X_test)

y_pred
```

```python
f"{np.mean(y_pred == y_test):.2f}"
```

```python
# もしくはこれ
knn.score(X_test, y_test)
```

### 解釈
- **手持ちのデータに対して** 97%の精度で正しい予測ができた
- これがビジネス要件を満たすかどうかは状況次第だけど、趣味プロダクトなら十分じゃないか？
